```python
import unicodedata
import sys

def shave_marks(txt):
    # Normalize to NFD (decompose characters into base + diacritics)
    norm_txt = unicodedata.normalize('NFD', txt)
    
    # Keep only characters that are not combining marks
    shaved = ''.join(
        c for c in norm_txt 
        if not unicodedata.combining(c)
        and 0 <= ord(c) <= sys.maxunicode   # ord() ensures valid code point
    )
    
    # Recompose back to NFC
    return unicodedata.normalize('NFC', shaved)
```

```python
print(shave_marks("café"))   # cafe
print(shave_marks("über"))   # uber
print(shave_marks("mañana")) # manana
```
